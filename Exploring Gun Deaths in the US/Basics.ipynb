# US Gun Deaths exploaration

## Reading the data

```python
import csv

file=open("guns.csv")
data=list(csv.reader(file))
print(data[:5])
```

## Removing the header

```python
headers=data[0]
data=data[1:]
print(headers)
print(data[:5])
```

## Deaths by year

```python
years=[row[1] for row in data]
years_counts={}

for year in years:
    
    if year in years_counts:
        years_counts[year]=years_counts[year]+1
    
    else:
        years_counts[year]=1
    

print(years_counts)
```

## Deaths by year and month

```python
import datetime as dt
dates=[ dt.datetime(year=int(row[1]), month=int(row[2]), day=1) for row in data]

print(dates[:5])
```

```python
date_counts={}

for date in dates:
    if date in date_counts:
        date_counts[date]= date_counts[date]+1
    
    else:
        date_counts[date]=1
    
date_counts
```

## Deaths by Gender

```python
sex_count={}
for row in data:
    
    if row[5] not in sex_count:
        sex_count[row[5]]=1
    
    sex_count[row[5]]=sex_count[row[5]]+1
sex_count
```

## Deaths by Race

```python
race_counts={}

for row in data:
    
    if row[7] not in race_counts:
        race_counts[row[7]]=1
    
    race_counts[row[7]] +=1

race_counts

    
    
```
